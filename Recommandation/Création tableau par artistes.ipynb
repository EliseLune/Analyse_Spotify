```python
def get_artists_to_list(i):
    res = all_tracks.iloc[i]['artists']
    res_list  = list(res[1:-1].split(","))
    final=[]
    for artist in res_list:
        if '(' in artist:
        else :
            if artist[0]=="'":
                final.append(artist[1:-1])
            if artist[0]==" ":
                final.append(artist[2:-1])
    return final
```

```python
for i in range(n):
    final = get_artists_to_list(i)
    liste_art.update (final)
```
