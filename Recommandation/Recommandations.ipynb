Il faut commencer par supprimer de notre base de données toutes les musiques déjà présentes dans notre playlist, et extraire les données dont on veut chercher des valeurs similaires :

```python
i = 0
playlist = [] #prendra l'énergie et l'accousticness de la musique
for i in range (0, len(table)) :
    playlist.append([table["energy"][i], table["acousticness"][i]])
    playlist_cible.drop(playlist_cible.loc[playlist_cible['id']==table.id[i]].index, inplace=True)
print(playlist)
```

```python
playlist_cible.head()
```

```python
nouvelle_playlist = pd.DataFrame({})
for i in range(0, len(playlist)):
    nouvelle_playlist = nouvelle_playlist.append(playlist_cible[(playlist_cible["energy"] > playlist[i][0] - 0.1) & (playlist_cible["energy"] < playlist[i][0] + 0.1) & (playlist_cible["acousticness"] > playlist[i][1] - 0.1) & (playlist_cible["acousticness"] < playlist[i][1] + 0.1)])
nouvelle_playlist = nouvelle_playlist.drop_duplicates() #enlever les doublons
nouvelle_playlist.head()
```

Dans la liste "cible" on ne peut mettre pour l'instant que deux critères, parmi acousticness, danseability, energy, instrumentalness, liveness, speechiness, valence, popularity.
pour mettre plus de liberté on peut faire ainsi : 

```python
#essaie de fonction plus générale :
cible = ["energy", "acousticness", "speechiness"]
i = 0
playlist = [] #prendra les données voulues de chaque musique
for i in range (0, len(table)) :
    playlist.append([table[cible[0]][i], table[cible[1]][i],table[cible[2]][i]])
    playlist_cible.drop(playlist_cible.loc[playlist_cible['id']==table.id[i]].index, inplace=True)
print(playlist)

nouvelle_playlist = pd.DataFrame({})

for i in range(0, len(playlist)):
    nouvelle_playlist = nouvelle_playlist.append(playlist_cible[(playlist_cible[cible[0]] > playlist[i][0] - 0.1) & (playlist_cible[cible[0]] < playlist[i][0] + 0.1)])
    for j in range(1, len(cible)) :
        nouvelle_playlist = nouvelle_playlist[(playlist_cible[cible[j]] > playlist[i][j] - 0.1) & (playlist_cible[cible[j]] < playlist[i][j] + 0.1)]
nouvelle_playlist = nouvelle_playlist.drop_duplicates() #enlever les doublons
nouvelle_playlist.head()
```
