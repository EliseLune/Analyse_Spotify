Il faut commencer par supprimer de notre base de données toutes les musiques déjà présentes dans notre playlist, et extraire les données dont on veut chercher des valeurs similaires :

```python
i = 0
playlist = [] #prendra l'énergie et l'accousticness de la musique
for i in range (0, len(table)) :
    playlist.append([table["energy"][i], table["acousticness"][i]])
    playlist_cible.drop(playlist_cible.loc[playlist_cible['id']==table.id[i]].index, inplace=True)
print(playlist)
```

```python
nouvelle_playlist = pd.DataFrame({})
playlist_cible = table #juste pour tester, aucun intérêt de faire ça en vrai
for i in range(0, len(playlist)):
    nouvelle_playlist = nouvelle_playlist.append(playlist_cible[(playlist_cible["energy"] > playlist[i][0] - 0.1) & (playlist_cible["energy"] < playlist[i][0] + 0.1) & (playlist_cible["acousticness"] > playlist[i][1] - 0.1) & (playlist_cible["acousticness"] < playlist[i][1] + 0.1)])
nouvelle_playlist = nouvelle_playlist.drop_duplicates() #enlever les doublons
print(nouvelle_playlist)
```

```python
#essaie de fonction plus générale :
cible = ["energy", "acousticness"]
i = 0
playlist = [] #prendra l'énergie et l'accousticness de la musique
for i in range (0, len(table)) :
    playlist.append([table[cible[0]][i], table[cible[1]][i]])
    playlist_cible.drop(playlist_cible.loc[playlist_cible['id']==table.id[i]].index, inplace=True)
print(playlist)

nouvelle_playlist = pd.DataFrame({})
playlist_cible = table #juste pour tester, aucun intérêt de faire ça en vrai
for i in range(0, len(playlist)):
    nouvelle_playlist = nouvelle_playlist.append(playlist_cible[(playlist_cible[cible[0]] > playlist[i][0] - 0.1) & (playlist_cible[cible[0]] < playlist[i][0] + 0.1) & (playlist_cible[cible[1]] > playlist[i][1] - 0.1) & (playlist_cible[cible[1]] < playlist[i][1] + 0.1)])
nouvelle_playlist = nouvelle_playlist.drop_duplicates() #enlever les doublons
print(nouvelle_playlist)
```

```python
#à partir des audiofeatures, pour faire même artiste + un paramètre similaire
donnee = "energy"
i = 0
playlist = [] #prendra l'artiste et la "donnée" de chaque musique
for i in range (0, len(table)) :
    playlist.append([table["artist"][i], table[donnee][i]])
    playlist_cible.drop(playlist_cible.loc[playlist_cible['id']==table.id[i]].index, inplace=True) #enlever de la playlist cible les musiques qu'on a déjà
print(playlist)

nouvelle_playlist = pd.DataFrame({})

for i in range(0, len(playlist)):
    nouvelle_playlist = nouvelle_playlist.append(playlist_cible[(playlist_cible["name"] == playlist[i][0]) & (playlist_cible[donne] > playlist[i][1] - 0.1) & (playlist_cible[donne] < playlist[i][1] + 0.1)])
nouvelle_playlist = nouvelle_playlist.drop_duplicates() #enlever les doublons
print(nouvelle_playlist)
```
